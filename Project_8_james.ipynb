```python
# Comma-separated values
breasts_tumor = pd.read_csv('data.csv')
breasts_tumor['cancer_or_not']= breasts_tumor['diagnosis']=='M'
breasts_tumor['cancer_or_not'] = breasts_tumor['cancer_or_not'].astype(int)


breasts_tumor.head()
```

```python



feature_cols=['radius_mean','texture_mean','perimeter_mean','area_mean']
x= breasts_tumor[feature_cols]
y= breasts_tumor.cancer_or_not

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
treeclf.fit(x, y)

graph = graphviz.Source(tree.export_graphviz(treeclf, out_file=None,
                                    feature_names=feature_cols,
                                    class_names=['0', '1', '2', '3'], filled = True))
svg = SVG(graph.pipe(format='svg'))
display(svg)


```
